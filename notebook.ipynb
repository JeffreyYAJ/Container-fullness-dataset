```


from ultralytics import YOLO

model = YOLO('yolov8n-cls.pt')

chemin_dataset = "/content/Container-fullness-dataset/"

# On lance la machine !
results = model.train(
    data=chemin_dataset,
    epochs=50,      # L'IA va lire tes dossiers 50 fois
    imgsz=224       # Taille standard pour la classification
)
```
